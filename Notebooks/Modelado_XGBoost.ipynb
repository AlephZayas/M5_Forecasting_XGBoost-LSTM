## Partición

```python
horizon = 28  # días

max_date = df['date'].max()
split_date = max_date - pd.Timedelta(days=horizon)

train = df[df['date'] <= split_date]
test = df[df['date'] > split_date]
```

## Definición de Variables

```python
# Unidad Muestral
um = ['id']
# Target multi-out

tgt = [f'target_{h}' for h in range(1, 29)]

# Variables discretas
vard = ['d_item_id', 'd_dept_id', 'd_cat_id', 'd_event_type_1']

# Variables Continuas
varc = ['lag_28', 'lag_35', 'lag_42', 'lag_56', 'rmean_28',
       'rstd_28', 'rmax_28', 'rmin_28', 'rmean_35', 'rstd_35', 'rmax_35',
       'rmin_35', 'rmean_42', 'rstd_42', 'rmax_42', 'rmin_42',
       'lag_35_rolling_28', 'lag_28_rolling_28', 'diff_28_35', 'diff_35_42',
       'price_lag_28', 'price_chg_28', 'price_max_year', 'norm_price',
       'price_std_28', 'store_sales_mean_lag_28',
       'cat_sales_mean_lag_28', 'dept_sales_mean_lag_28', 'wday_sin',
       'wday_cos', 'month_sin', 'month_cos', 'week', 'week_sin', 'week_cos']

# Variables Discretas
vard = ['d_item_id', 'd_dept_id', 'd_cat_id', 'd_event_type_1',
        'snap_active','snap_event']

feat = vard + varc 
```

```python
X_train = train[feat].copy()
y_train = train[tgt].copy()

X_test = test[feat].copy()
y_test = test[tgt].copy()
```

```python
base_model = xgb.XGBRegressor(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=8,
    subsample=0.8,
    colsample_bytree=0.8,
    tree_method="hist",
    random_state=42,
    n_jobs=-1,
    objective='reg:tweedie'
)

model = MultiOutputRegressor(base_model)
```

```python
model.fit(X_train, y_train)
```

## Predicción y Evaluación

```python
y_pred = model.predict(X_test)
```

```python
rmse = np.sqrt(mean_squared_error(test[tgt],y_pred) )
mae = mean_absolute_error(test[tgt], y_pred)
mape = mean_absolute_percentage_error(test[tgt], y_pred)

print(f'RMSE: {rmse}')
print(f'MAE: {mae}')
print(f'MAPE: {mape}')
```

```python
def compute_rmsse_denominator(df):
    denom = {}

    for uid, group in df.groupby('id'):
        y = group['sales'].values
        
        # diferencias lag 1
        diff = np.diff(y)
        
        # evitar división por cero
        denom[uid] = np.mean(diff**2) if len(diff) > 0 else 0.0

    return denom

def rmsse_score(y_true, y_pred, ids, denom_dict):
    errors = []

    for i, uid in enumerate(ids):
        denom = denom_dict.get(uid, 0)

        # evitar división por cero
        if denom == 0:
            continue

        mse = np.mean((y_true[i] - y_pred[i])**2)
        rmsse = np.sqrt(mse / denom)

        errors.append(rmsse)

    return np.mean(errors)

```

```python
denom = compute_rmsse_denominator(train)
```

```python
y_true = test[tgt].values
ids = test['id'].values

score = rmsse_score(y_true, y_pred, ids, denom)

print("RMSSE:", score)
```
