```python
import pandas as pd
import duckdb

pd.set_option('display.max_columns', 200)
```

```python
# Conectamos a la base 

con = duckdb.connect(r'D:\Diplomado\TecnicasCognitivasIntroduccionABD\Proyecto\DataCleaned\m5.db')

con.execute(
    "create schema if not exists m5_schema"
)
```

```python
con.sql(
    """
    create table m5_schema.df_calendar as (
        select *
        from read_csv("D:\Diplomado\TecnicasCognitivasIntroduccionABD\Proyecto\DataCleaned\df_calendar.csv")
    )
    """
)
```

```python
con.sql(
    """
    create table m5_schema.df_sales_train_val as (
        select *
        from read_csv("D:\Diplomado\TecnicasCognitivasIntroduccionABD\Proyecto\DataCleaned\df_sales_train_val.csv")
    )
    """
)
con.sql(
    """
    create table m5_schema.df_sell_prices as (
        select *
        from read_csv("D:\Diplomado\TecnicasCognitivasIntroduccionABD\Proyecto\DataCleaned\df_sell_prices.csv")
    )
    """
)
```

```python
print(con.sql(
    """
    select *
    from df_calendar
    limit 3;
    """
))

print(con.sql(
    """    
    select *
    from df_sales_train_val
    limit 3;
    """
))

print(
    con.sql(
    """
        select *
        from df_sell_prices
        limit 3;
    """
    )
)
```

```python
# Vemos cuántos años tiene el histórico

con.sql(
    """
    select
        min(date) as min_date,
        max(date) as max_date
    from df_calendar;
    """
)
```

```python
# Limitamos la historia a dos años
con.sql(
    """
    create table df_calendar_filtered as
    select *
    from df_calendar
    where date >= '2014-06-19'
    """
)

con.sql(
    """
    select * 
    from df_calendar_filtered
    limit 3
    """
)
```

```python
# Unimos 

con.sql(
    """
    create table df as
    select *
    from df_sales_train_val A
    inner join df_calendar_filtered B
    on A.day = B.d
    """
)
con.sql(
    """
    summarize df
    """
)
```

```python
con.sql(
    """
    select *
    from df_sell_prices
    limit 3
    """
)
```

```python
con.sql(
    """
    alter table df_sell_prices
    add column master_id varchar
    """
)

con.sql(
    """
    update df_sell_prices
    set master_id = concat(item_id, '_', store_id, '_', wm_yr_wk)
    """
)

con.sql(
    """
    select * from df_sell_prices limit 3
    """
)
```

```python
# Creamos el master_id en df para poder hacer el join con df_sell_prices

con.sql(
    """
    alter table df
    add column master_id varchar
    """
)

con.sql(
    """
    update df
    set master_id = replace(id, '_validation', '') || '_' || cast(wm_yr_wk as varchar)
    """
)

con.sql(
    """
    select * from df limit 3;
    """
)
```

```python
# Terminamos el join

con.sql(
    """
    create table df_sell_prices_master as
    select sell_price, master_id
    from df_sell_prices
    """
)
```

```python
con.sql(
    """
    create table df_full as 
    select *
    from df A
    left join df_sell_prices_master B
    on A.master_id = B.master_id
    """
)
```

```python
# Guardamos el resultado

con.sql(
    """
    select *
    from df_full
    """
).write_parquet(r"D:\Diplomado\TecnicasCognitivasIntroduccionABD\Proyecto\DataCleaned\df_full.parquet")
```
